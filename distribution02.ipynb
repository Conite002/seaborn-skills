```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('datas/StudentsPerformance.csv')
df.head()
```

```python
plt.figure(figsize=(10,3), dpi=150)
sns.boxplot(data=df,y="reading score",x="parental level of education",hue="test preparation course")
```

```python
plt.figure(figsize=(10,5), dpi=150)
sns.violinplot(data=df, y="parental level of education", x="reading score", hue="test preparation course")
```

```python
plt.figure(figsize=(10,5), dpi=150)
sns.violinplot(data=df, y="parental level of education", x="reading score",hue="test preparation course",split=True, bw=0.2)
```

```python
plt.figure(figsize=(5,3),dpi=200)
sns.swarmplot(data=df, x="math score",size=2, y='gender',hue="test preparation course")
plt.legend(bbox_to_anchor=(1.35,0.5))
```

# Boxplot or Boxenplot

```python
sns.boxenplot(x='math score', y='test preparation course', data=df, hue='gender')
```

```python
sns.boxplot(data=df, x='math score',y='test preparation course')
```

```python
sns.boxplot(data=df, x='math score',y='test preparation course', hue="gender")
```

# Heatmap(linewith, annot=True, cmap="viridis")

```python
df=pd.read_csv("datas/country_table.csv", index_col="continent")
df = df.groupby(by='continent').sum()
df
```

```python
plt.figure(dpi=150)
sns.heatmap(data=df,linewidths=0.5, annot=True)
```

```python
plt.figure(dpi=150)
sns.clustermap(data=df,linewidths=0.5, annot=True, cmap='viridis')
```
