```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("datas/dm_office_sales.csv")
df.head()
```

```python
plt.figure(figsize=(7,5), dpi=150)
sns.rugplot(x="salary", data=df, height=0.5)
```

```python
plt.figure(dpi=200,figsize=(8,4))
sns.set(style="darkgrid")
sns.displot(data=df,x="salary",bins=10,ls ="-",kde=True,hue="level of education")
plt.savefig("figures/displot.png")
```

```python
plt.figure(dpi=200,figsize=(8,4))
sns.set(style="darkgrid")
sns.displot(data=df,x="salary",bins=10,palette="RPGn", lw=10,edgecolor="blue",ls ="-",kde=True, rug=True)
plt.savefig("figures/displot02.png")
```

```python
sns.kdeplot(data=df,x="salary", hue="level of education", marker='+')
```

```python
sns.kdeplot(data=df,x="salary", hue="level of education",linewidth=.5, marker='o',markersize=5)
```

```python
sns.rugplot(data=df,x="salary")
```

```python
np.random.seed(41)
sample_ages = np.random.randint(0,100,200)
sample_ages = pd.DataFrame(data=sample_ages,columns=["ages"])
```

```python
sns.rugplot(data=sample_ages,x="ages",linewidth=5, color="red")
```

```python
sns.displot(data=sample_ages,x="ages", rug=True, bins=30)
```

```python
sns.displot(data=sample_ages,x="ages", rug=True, bins=30, kde=True)
```

```python
sns.kdeplot(data=sample_ages,x="ages", clip=[0,100],bw_adjust=0.5, fillstyle='bottom')
```

```python
sns.kdeplot(data=sample_ages,x="ages", clip=[-20,100],bw_adjust=0.5,shade=True)
```
