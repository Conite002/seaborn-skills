```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("dm_office_sales.csv")
df.head()
```

```python
df.corr()
```

```python
plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x="salary", y="sales", data=df, hue="level of education")
plt.savefig("seaborn_fig01.png")
```

```python
plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x="salary", y="sales", data=df, hue="training level")
plt.savefig("seaborn_fig02.png")
```

```python
plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x="salary", y="sales", data=df, hue="training level", palette="PRGn")
plt.savefig("seaborn_fig03.png")
```

```python
plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x="salary", y="sales", data=df, hue="training level", palette="Dark2")
plt.savefig("seaborn_fig04.png")
```

```python
plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x="salary", y="sales", data=df,size="salary")
plt.savefig("seaborn_fig05.png")
```

```python
plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x="salary", y="sales", data=df, s=100, alpha=0.2)
plt.savefig("seaborn_fig06.png")
```

```python
plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x="salary", y="sales", data=df, s=100,style="level of education", hue="level of education")
plt.savefig("seaborn_fig07.png")
```
